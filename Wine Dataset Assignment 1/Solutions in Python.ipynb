```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_table("wine.data",header=None,delimiter=",")
```

```python
Y = np.asarray(df[0])
print(Y[0:10])
```

```python
X = np.asmatrix(df)
print(X[:5,:])
```

```python
means = [0] 
stds = [1]

#Calculating mean and standard deviation to perform batch normalization
for i in range(13):
    means.append(np.mean(X[:,i]))
    stds.append(np.std(X[:,i]))
    X[:,i] = (X[:,i] - means[-1])/stds[-1]
    
print(means)
print(stds)
print(X[:5,:])
```

# TRAINING THE MODEL

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_cv, Y_train, Y_cv = train_test_split(X,Y,train_size=0.75,test_size=0.25,random_state=12,stratify=Y)
```

```python
print(np.shape(X_train))
print(np.shape(X_cv))
print(np.shape(Y_train))
print(np.shape(Y_cv))
```

### Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
```

```python
model1 = LogisticRegression(tol=0.00001,max_iter=10000,multi_class='multinomial',n_jobs=2,verbose=5,solver='lbfgs')
```

```python
model1.fit(X_train,Y_train)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
accuracy_score(model1.predict(X_train),Y_train)
```

```python
accuracy_score(model1.predict(X_cv),Y_cv)
```

```python
confusion_matrix(model1.predict(X_train),Y_train)
```

```python
confusion_matrix(model1.predict(X_cv),Y_cv)
```

### Gradient Boosted Trees

```python
import xgboost as xgb
```

```python
model2 = xgb.XGBClassifier(max_depth=5,silent=False,reg_alpha = 1,reg_lambda=0,seed=124)
```

```python
model2.fit(X_train,Y_train)
```

```python
accuracy_score(model2.predict(X_train),Y_train)
```

```python
accuracy_score(model2.predict(X_cv),Y_cv)
```

```python
confusion_matrix(model2.predict(X_cv),Y_cv)
```

### Voting Classifier

```python
from sklearn.ensemble import VotingClassifier
```

```python
model3 = VotingClassifier(estimators=[('logistic',model1),('xgb',model2)],voting='soft')
```

```python
model3.fit(X_train,Y_train)
```

```python
accuracy_score(model3.predict(X_train),Y_train)
```

```python
accuracy_score(model3.predict(X_cv),Y_cv)
```

```python
confusion_matrix(model3.predict(X_cv),Y_cv)
```
